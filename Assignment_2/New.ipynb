```python
from keras.utils import to_categorical
```

```python
t = df.values
np.random.shuffle(t)
print(t.shape)

Y = np.zeros((t.shape[0], 1))
print(Y.shape)
for i in range(8):
    Y = np.append(Y, to_categorical(t[:, i]), axis = 1)
for i in range(8, 18):
    y = t[:, i].reshape(-1,1)
    y = (y-np.min(y))/float(np.max(y) - np.min(y))
    Y = np.append(Y, y, axis = 1)
for i in range(18, 22):                                       #Not considering feature 23 since all are distinct
    Y = np.append(Y, to_categorical(t[:, i]), axis = 1)

#y = t[:, 23].reshape(-1,1)
#y = (y-np.min(y))/float(np.max(y) - np.min(y))
#Y = np.append(Y, y, axis = 1)

Y = Y[:, 1:]
Y.shape
```

```python
print(Y[:,0])
Yn = Y[:, 1:]
Yn[0]
Y[:,1418]
```

```python
x_tr = Y[:800, :]
x_ts = Y[800:, :]

y_tr = to_categorical(t[:, 24])[:800, :]
y_ts = to_categorical(t[:, 24])[800:, :]

print(x_tr.shape)
print(x_ts.shape)
print(y_tr.shape)
print(y_ts.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(units=1152, activation='relu'),
    keras.layers.Dense(units=576, activation='relu'),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dense(units=18, activation='relu')
])
```

```python
model.compile(optimizer='SGD', 
              loss=tf.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(
    x_tr, 
    y_tr,
    epochs=30, 
    steps_per_epoch=100,
    #validation_data=val_dataset.repeat(), 
    #validation_steps=2
)
```

```python
model.evaluate(x_ts, y_ts)
```
